```python
import pandas as pd
```

```python
df = pd.read_excel('Employee.xlsx')
df.head()
```

```python
if "Unnamed: 0" in df.columns:
    df = df.set_index("Unnamed: 0")

df.insert(0, "Unnamed", [f"Unnamed {i}" for i in range(len(df))])
df.head()
```

```python
df = df.set_index('Name')
df.head()
```

```python
x = lambda x: [i.str.replace(" ", "_") if len(i) != 1 else i for i in df.columns]
print(x)
```
